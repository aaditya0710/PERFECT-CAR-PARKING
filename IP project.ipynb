# CORRECT CAR PARKING

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
main_img = cv2.imread("JUPYTER_DATA/Car_top1.jpg")
plt.imshow(main_img)
```

```python
img = cv2.imread("JUPYTER_DATA/Car_top1.jpg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
pts = []
for aperturesize in [5,7]:
    edges = cv2.Canny(gray,25,100,apertureSize = aperturesize)
        
    lines = cv2.HoughLines(edges,1,np.pi/180,30)
    for rho,theta in lines[0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        cv2.line(img,(x1,y1),(x2,y2),(255,0,0),2)
        pts.append(x1)
```

```python
plt.imshow(img)
```

```python
blurred = cv2.blur(main_img,(10,10))
edges = cv2.Canny(blurred,50,60)
plt.imshow(edges)
```

```python
arr=[]
for i in range(edges.shape[1]):
    for j in range(edges.shape[0]):
        if edges[j][i]!=0:
            arr.append(i)
```

```python
if max(pts)>=max(arr) and min(pts)<=min(arr):
    print("correctly parked")
else:
    print("incorrectly parked")
```
